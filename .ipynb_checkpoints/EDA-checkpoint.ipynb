### Performing questions asked in the task

### 1. Identify null values in the data set and suggest/implement methods to replace them

```python
missing_values = df.isnull().sum()
print(missing_values)
```

**Inference**: From above we can observe that **Cabin** column has most no of missing values followed by **Age** and **Embarked** respectively.

### Strategy for handling missing values
- We will use median to fill missing values of numerical columns as it will be robust to outliers
- We will use mode to fill missing values of categorical columns as it will preserve common values

```python
df.fillna({
    "Age": df["Age"].median(),
    "Embarked": df["Embarked"].mode()[0],    #we are selecting first mode value incase there are multiple modes
    "Fare": df["Fare"].median(), 
    "Cabin": "Unknown"  # Replacing missing Cabin values as "Unknown" as there are mix of complex values for median and mode strategy
}, inplace=True)
```

```python
df.head(10)
```

**Inference**: From above we can observe that missing values has been replaced with normalized value from same column using median and mode strategy.

### 3. Identify outliers, if any, for all features

```python
for col in df.select_dtypes(include=["int64", "float64"]).columns:
    plt.figure(figsize=(10, 7))
    sns.boxplot(x=df[col], color="skyblue")
    plt.title(f"Box Plot of {col}")
    plt.show()
```

**Inference**: Here we have visualized outliers using box plot and as a result **Fare** Column has most no of outliers.

### Strategy for handling missing values
- Here we will use **Z-Score** technique to handle outliers as it works well with large dataset and doesn't cause data loss

```python
def remove_outliers(df, threshold=3):
    numerical_cols = ["Fare", "Age"]
    df_filtered = df[(np.abs(zscore(df[numerical_cols])) < threshold).all(axis=1)]
    return df_filtered


print("Dataset Size Before Outlier Removal:", df.shape)

# Remove outliers
df = remove_outliers(df)

print("Dataset Size After Outlier Removal:", df.shape)
```

**Inference:** From above we can observed that using **Z-score** technique **27** outliers were removed

### 4. Draw correlation plot for the numeric features in the dataset.

```python
#filter numercal columns
numeric = df.select_dtypes(include=["float64", "int64"])

#correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(numeric.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

- From corelation plot we can conclude that features with values close to **+1** has strong correlation
- From corelation plot we can conclude that features with values close to **-1** has inverse correlation
- From corelation plot we can conclude that features with values close to **0** has no correlation

- Hence Age has positive correlation with Fare and played role deciding fare amount 
- Passenger Class and Fare are negatively correlated with each other

### 7. Identify strategies for improving execution performance

- Used median/mode instead of mean to handle missing values (robust to outliers).
- Removed outliers using Z-score to ensure a more balanced dataset.
- Used `.select_dtypes()` to filter numerical columns efficiently.
- Minimized redundant calculations using inplace=True wherever possible
- Instead of looping over each column separately in Z-score technique, used `.all(axis=1)` as it checks all columns row-wise.
